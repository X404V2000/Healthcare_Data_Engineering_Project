```python
import psycopg2
import .dbconn
```

```python
try:
    conn=psycopg2.connect(
        host=.dbconn.host,
        dbname=.dbconn.dbname,
        user=.dbconn.user,
        port=.dbconn.port,
        password=.dbconn.password
    )
    print('connected...')
except Exception as e:
    print(e)
```

```python
conn.autocommit=True
```

```python
cur=conn.cursor()
```

```python
cur.execute("SET search_path TO your_schema_name, public")
```

```python
def load_data():
    try:
        cur.execute(
            """COPY healthcare_org.encounters(patient_id,date,diagnosis,physician,cost,hospital)
                FROM '/encounters.csv'
                DELIMITER ','
                CSV HEADER;
            """)
        print('Data loading complete...')
    except Exception as e:
        return f"{e}"
load_data()
```
